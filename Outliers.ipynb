```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('Ventas_totales.csv') 
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data['salon_ventas']=data['salon_ventas'].fillna(round(data['salon_ventas'].mean(),1))
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos con promedio o media de la variable subtotal_ventas_alimentos_bebidas
data["subtotal_ventas_alimentos_bebidas"]=data["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos con promedio o media de la variable otros
data["otros"]=data["otros"].fillna(round(data["otros"].mean(),1))
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#A continuación se sustituiran los valores nulos con mediana de las variables
#tarjetas_credito,almacen,indumentaria_calzado_textiles_hogar  e electronicos_articulos_hogar     
data["tarjetas_credito"]=data["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(),1))
data

data["almacen"]=data["almacen"].fillna(round(data["almacen"].median(),1))
data

data["indumentaria_calzado_textiles_hogar"]=data["indumentaria_calzado_textiles_hogar"].fillna(round(data["indumentaria_calzado_textiles_hogar"].median(),1))
data

data["electronicos_articulos_hogar"]=data["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Las columnas que contienen valores nulos se eliminaron para un mayor manejo de datos y nuestro dataframe
#quedó con 16 variables
data2= data.drop(['alimentos_preparados_rotiseria','verduleria_fruteria','carnes','lacteos','panaderia','bebidas','otros_medios','tarjetas_debito'], axis=1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Ahora visualizaremos si nuestras variables con outliers y si es así aplicaremos diferentes metodos 
#para lidiar con esos valores
boxplot = data2.boxplot(column=['ventas_precios_constantes','salon_ventas','tarjetas_credito','subtotal_ventas_alimentos_bebidas','indumentaria_calzado_textiles_hogar','electronicos_articulos_hogar']) 
```
